```python
def CountCars(image):
    '''Takes Input of path of the image name and provides number of cars in the image'''
    #Read all packages
    import cv2
    import matplotlib.pyplot as plt
    import cvlib as cv
    from cvlib.object_detection import draw_bbox
    from skimage.io import imread, imshow
    
    
    # Read image
    image = imread(str(image))
    bbox, label, conf = cv.detect_common_objects(image)
    output_image = draw_bbox(image, bbox, label, conf)
    plt.imshow(output_image)
    return plt.show(), 'Number of cars in the image are: ' + str(label.count('car'))
    
```

```python
CountCars('Images/MoreCars.jpg')
```
